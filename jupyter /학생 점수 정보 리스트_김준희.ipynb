## 학생 정보를 저장할 students라는 빈 리스트를 먼저 만들고 앞으로 여기에 학생 한 명 한 명의 정보를 작은 리스트로 담을 예정입니다

```python
students =[]
```

## range(len(...))를 써서 이름 리스트를 하나씩 돌며, 이름과 점수를 짝지어 students에 집어넣어 중첩 리스트 구조를 썼습니다

```python
new_names = ["김준희","김철수","홍길동"]
new_scores = [90,85,95]

for i in range (len(new_names)):
    name = new_names[i]
    score = new_scores[i]
    students.append([name,score]) 

print("추가 완료 : ", students)
```

## for문으로 리스트를 훑으면서, if문으로 내가 찾는 이름인지 확인하고 맞다면 그 줄의 점수 칸(s[1])만 딱 바꿨스비다

```python
target = "김철수"
new_val = 100

for s in students:
    if s[0] == target:
        s[1] = new_val
        print(f"{target}의 점수를 {new_val}로 수정했습니다.")
print("수정 후 목록: ", students)
```

## students[i][1]은 i번째 줄의 이름표를 확인하는 건데 이름이 맞으면 del로 그 칸을 날려버립니다.

```python
delete_name = "김준희"

for i in range(len(students)):
    if students[i][1] == delete_name:
        del students[i]
        print(f"{delete_name}의 정보를 삭제했습니다.")
print("삭제 후 목록: ", students)        
```

## 리스트 컴프리헨션을 써서 복잡한 반복문 없이 점수만 싹 모았습니다 그다음 max, sum 같은 파이썬 기본 도구로 통계를 냈스비ㅏㄷ.

```python
scores= [s[1] for s in students]

max_score = max (scores)
min_score = min (scores)
avg_score = sum (scores) / len (scores)

print(f"최고점: {max_score}, 최저점: {min_score}, 평균점: {avg_score}")
```
