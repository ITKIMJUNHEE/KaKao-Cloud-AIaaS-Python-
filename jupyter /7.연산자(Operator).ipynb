```python
a = 10
b = 3

# 덧셈 연산자(+)
print(f"덧셈: {a} + {b} = {a + b}") # 결과: 13

# 뺄셈 연산자(-)
print(f"뺄셈: {a} - {b} = {a - b}") # 결과: 7

# 곱셈 연산자(*)
print(f"곱셈: {a} * {b} = {a * b}") # 결과: 30
```

```python
# 나눗셈 연산자(/) - 실수 결과 반환
print(f"나눗셈: {a} / {b} = {a / b}") # 결과: 3.3333...

# 정수 나눗셈 연산자(//) - 몫만 반환
print(f"정수 나눗셈: {a} // {b} = {a // b}") # 결과: 3

# 나머지 연산자(%)
print(f"나머지: {a} % {b} = {a % b}") # 결과: 1

# 거듭제곱 연산자(**)
print(f"거듭제곱: {a} ** {b} = {a ** b}") # 결과: 1000
```

```python
# 문자열과 덧셈 연산자
str1 = "Hello"
str2 = "World"
print(f"문자열 연결: {str1} + {str2} = {str1 + str2}") # 결과: HelloWorld

# 문자열과 곱셈 연산자
print(f"문자열 반복: {str1} * 3 = {str1 * 3}") # 결과: HelloHelloHello
```

```python
# 비교 연산자 예시
x = 10
y = 5
z = 10
```

```python
# 동등 비교 연산자(==)
print(f"{x} == {y} = {x == y}") # 결과: False
print(f"{x} == {z} = {x == z}") # 결과: True

# 부동 비교 연산자(!=)
print(f"{x} != {y} = {x != y}") # 결과: True

# 크다(>)
print(f"{x} > {y} = {x > y}")   # 결과: True

# 작다(<)
print(f"{x} < {y} = {x < y}")   # 결과: False
```

```python
# 크거나 같다(>=)
print(f"{x} >= {z} = {x >= z}")  # 결과: True

# 작거나 같다(<=)
print(f"{x} <= {y} = {x <= y}")  # 결과: False

# 문자열 비교 (알파벳 순서로 비교)
str1 = "apple"
str2 = "banana"
print(f"'{str1}' < '{str2}' = {str1 < str2}")  # 결과: True (a가 b보다 알파벳 순서상 앞에 있음)

# 다른 타입 비교
print(f"3 == '3' = {3 == '3'}")  # 결과: False (타입이 다름)
print(f"3 == int('3') = {3 == int('3')}")  # 결과: True (타입 변환 후 비교)
```

```python
# fruits 집합을 먼저 정의
fruits = {"사과", "바나나", "체리", "오렌지"} # 중괄호를 사용하여 집합 생성
# for 루프와 함께 사용
for fruit in fruits:
    if fruit in ["사과", "오렌지"]:
        print(f"{fruit}는 과일 주스로 적합합니다.")

# 조건문과 함께 사용
user_input = "체리"
if user_input in fruits:
    print(f"{user_input}는 과일 목록에 있습니다.")
else:
    print(f"{user_input}는 과일 목록에 없습니다.")
```

```python
# 집합에서의 활용 (매우 빠른 검색 - O(1) 시간 복잡도)
colors = {"빨강", "파랑", "녹색"}
print(f"'파랑' in colors: {'파랑' in colors}") # 결과: True
print(f"'노랑' in colors: {'노랑' in colors}") # 결과: False

# 딕셔너리에서의 활용 (키 검색)
user = {"name": "홍길동", "age": 30, "email": "hong@example.com"}
print(f"'name' in user: {'name' in user}") # 결과: True (키 존재)
print(f"'address' in user: {'address' in user}") # 결과: False (키 없음)
print(f"'홍길동' in user: {'홍길동' in user}") # 결과: False (값은 검색 안됨)
print(f"'홍길동' in user.values(): {'홍길동' in user.values()}") # 결과: True (값 검색)
```

```python
# 여러 컬렉션의 교차 확인
requested_toppings = ["버섯", "치즈", "피망"]
available_toppings = ["치즈", "올리브", "피망", "페퍼로니", "버섯"]

for topping in requested_toppings:
    if topping in available_toppings:
        print(f"{topping} 추가")
    else:
        print(f"{topping}은(는) 재고가 없습니다.")
```

```python
# 논리 연산자 예시
a = True
b = False
c = True
```

```python
# and 연산자 (두 조건이 모두 참일 때만 참)
print(f"{a} and {b} = {a and b}")  # 결과: False
print(f"{a} and {c} = {a and c}")  # 결과: True

# or 연산자 (두 조건 중 하나라도 참이면 참)
print(f"{a} or {b} = {a or b}")    # 결과: True
print(f"{b} or {b} = {b or b}")    # 결과: False

# not 연산자 (조건의 결과를 반대로)
print(f"not {a} = {not a}")        # 결과: False
print(f"not {b} = {not b}")        # 결과: True

# 복합 논리 연산
print(f"{a} and {b} or {c} = {a and b or c}")      # 결과: True
print(f"{a} and ({b} or {c}) = {a and (b or c)}")  # 결과: True
```

```python
# 정수 값의 XOR
a = 5     # 이진수: 101
b = 3     # 이진수: 011
result = a ^ b # 이진수: 110 (십진수: 6)
print(result) # 출력: 6
```

```python
True ^ True   # False (같은 값)
False ^ False # False (같은 값)
True ^ False  # True (다른 값)
False ^ True  # True (다른 값)
```

```python
# XOR의 실용적 활용 예시
# 두 값이 서로 다른지 확인하기
has_key = True
has_permission = False

# 키를 가지고 있거나 권한이 있지만, 둘 다 있거나 둘 다 없으면 안 됨
can_access = has_key ^ has_permission
print(f"키 또는 권한 중 하나만 있을 때 접근 가능: {can_access}") # 결과: True
```

```python
def xor(x, y):
    return (x or y) and not (x and y)

a = True
b = False
c = True

print(f"xor({a}, {b}) = {xor(a, b)}") # 결과: True
print(f"xor({a}, {c}) = {xor(a, c)}") # 결과: False
print(f"xor({b}, {b}) = {xor(b, b)}") # 결과: False
```

```python
# 기본 할당 연산자(=)
x = 10
print(f"x = {x}") # 결과: 10

# 복합 할당 연산자(+=)
x += 5  # x = x + 5와 동일
print(f"x += 5 -> x = {x}") # 결과: 15

# 복합 할당 연산자(-=)
x -= 3  # x = x - 3와 동일
print(f"x -= 3 -> x = {x}") # 결과: 12
```

```python
# 복합 할당 연산자(*=)
x *= 2  # x = x * 2와 동일
print(f"x *= 2 -> x = {x}") # 결과: 24

# 복합 할당 연산자(/=)
x /= 4  # x = x / 4와 동일
print(f"x /= 4 -> x = {x}") # 결과: 6.0

# 복합 할당 연산자(//=)
x //= 2 # x = x // 2와 동일
print(f"x //= 2 -> x = {x}") # 결과: 3.0
```

```python
# 복합 할당 연산자(%=)
x %= 2  # x = x % 2와 동일
print(f"x %= 2 -> x = {x}") # 결과: 1.0

# 복합 할당 연산자(**=)
y = 2
y **= 3 # y = y ** 3와 동일
print(f"y **= 3 -> y = {y}") # 결과: 8
```

```python
# 문자열에 대한 복합 할당 연산자
message = "Hello"
message += " " + "World" # message = message + " " + "World"와 동일
print(f"message += ' World' -> message = {message}") # 결과: Hello World

# 다중 할당
a, b, c = 1, 2, 3
print(f"a, b, c = 1, 2, 3 -> a={a}, b={b}, c={c}") # 결과: a=1, b=2, c=3

# 값 교환(swap)
a, b = 10, 20
print(f"교환 전: a={a}, b={b}")
a, b = b, a
print(f"교환 후: a={a}, b={b}") # 결과: a=20, b=10
```

```python
# 기본 우선순위
result = 2 + 3 * 4
print(f"2 + 3 * 4 = {result}") # 결과: 14 (곱셈이 덧셈보다 우선순위가 높음)

# 괄호를 사용한 우선순위 변경
result = (2 + 3) * 4
print(f"(2 + 3) * 4 = {result}") # 결과: 20 (괄호 안의 연산이 먼저 수행됨)

# 복합 연산에서의 우선순위
result = 2 ** 3 * 4 + 5
print(f"2 ** 3 * 4 + 5 = {result}") # 결과: 37 (2^3=8, 8*4=32, 32+5=37)

# 논리 연산자 우선순위
# not > and > or 순으로 우선순위가 높음
result = True or False and not True
print(f"True or False and not True = {result}") # 결과: True
```

```python
# 거듭제곱 연산자는 우->좌 결합성을 가짐
result = 2 ** 3 ** 2
print(f"2 ** 3 ** 2 = {result}") # 결과: 512

# 계산 과정: 2 ** (3 ** 2) = 2 ** 9 = 512
# 만약 좌->우 결합성이었다면: (2 ** 3) ** 2 = 8 ** 2 = 64

# 할당 연산자도 우->좌 결합성을 가짐
a = b = c = 10
print(f"a = b = c = 10 결과: a={a}, b={b}, c={c}")
# 계산 과정: a = (b = (c = 10))

# 거듭제곱 vs 단항 연산자 우선순위와 결합성
result = -2 ** 2
print(f"-2 ** 2 = {result}") # 결과: -4

# 계산 과정: -(2 ** 2) = -4
# 거듭제곱 연산자(**)가 단항 연산자(-)보다 우선순위가 높음
```
