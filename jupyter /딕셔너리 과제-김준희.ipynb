```python
address_book = {}
```

```python
def add_contact():
    name = input("추가할 이름을 입력하세요: ")
    if name in address_book:
        print(f">> 에러: '{name}'님은 이미 존재합니다.")
        return

    phone = input("전화번호: ")
    email = input("이메일: ")
    address = input("주소: ")
    
    address_book[name] = {
        "phone": phone,
        "email": email,
        "address": address
    }
    print(f"\n>> '{name}'님의 연락처가 저장되었습니다.")

def delete_contact():
    name = input("삭제할 이름을 입력하세요: ")
    if name in address_book:
        del address_book[name]
        print(f">> '{name}'님의 연락처를 삭제했습니다.")
    else:
        print(f">> 삭제 실패: '{name}'님을 찾을 수 없습니다.")

def search_contact():
    name = input("검색할 이름을 입력하세요: ")
    info = address_book.get(name)
    
    if info:
        print(f"\n[{name}] 검색 결과:")
        print(f" - 전화번호: {info['phone']}")
        print(f" - 이메일: {info['email']}")
        print(f" - 주소: {info['address']}")
    else:
        print(f">> '{name}'님은 주소록에 없습니다.")

def edit_contact():
    name = input("수정할 이름을 입력하세요: ")
    if name in address_book:
        print(f"\n--- [{name}] 정보 수정 ---")
        print("(변경하지 않으려면 엔터를 치세요)")
        
        new_phone = input(f"새 전화번호 (기존: {address_book[name]['phone']}): ")
        new_email = input(f"새 이메일 (기존: {address_book[name]['email']}): ")
        new_address = input(f"새 주소 (기존: {address_book[name]['address']}): ")
        
        if new_phone: address_book[name]['phone'] = new_phone
        if new_email: address_book[name]['email'] = new_email
        if new_address: address_book[name]['address'] = new_address
        
        print(f">> '{name}'님의 정보가 수정되었습니다.")
    else:
        print(f">> 수정 실패: '{name}'님을 찾을 수 없습니다.")

def show_all():
    print(f"\n{'='*15} 주소록 전체 보기 {'='*15}")
    if not address_book:
        print("현재 저장된 연락처가 없습니다.")
    else:
        for name, info in address_book.items():
            print(f"이름: {name}")
            print(f"  전화: {info['phone']} | 메일: {info['email']} | 주소: {info['address']}")
            print("-" * 45)
    print(f"{'='*46}")
```

```python
while True:
    print("\n--- 파이썬 주소록 관리 프로그램 ---")
    print("1. 추가  2. 삭제  3. 검색  4. 수정  5. 전체보기  6. 종료")
    
    choice = input("메뉴 번호를 선택하세요: ")
    
    if choice == "1":
        add_contact()
    elif choice == "2":
        delete_contact()
    elif choice == "3":
        search_contact()
    elif choice == "4":
        edit_contact()
    elif choice == "5":
        show_all()
    elif choice == "6":
        print("프로그램을 종료합니다.")
        break
    else:
        print("잘못된 입력입니다. 다시 선택해주세요.")
```
