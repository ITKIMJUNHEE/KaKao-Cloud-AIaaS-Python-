```python
# 실생활 예제: 온라인 쇼핑몰 할인 판단
total_price = 55000  # 구매 금액
discount_threshold = 50000 # 할인 기준액

# 할인 조건 검사
if total_price > discount_threshold:
    print(f"구매금액 {total_price}원은 할인 기준액({discount_threshold}원)을 초과했습니다.")
    print("5% 할인이 적용됩니다!")
    discount = total_price * 0.05
    final_price = total_price - discount
    print(f"할인금액: {discount:.0f}원")
    print(f"최종 결제금액: {final_price:.0f}원")

print("이용해 주셔서 감사합니다.") # 조건과 무관하게 항상 실행
```

```python
# 실생활 예제: 합격/불합격 판단
exam_score = 68
pass_threshold = 70

if exam_score >= pass_threshold:
    print(f"축하합니다! {exam_score}점으로 합격입니다.")
else:
    points_needed = pass_threshold - exam_score
    print(f"아쉽게도 {exam_score}점으로 불합격입니다.")
    print(f"합격까지 {points_needed}점이 부족합니다.")
```

```python
def grade_calculator(score):
    if score >= 90:
        return "A"
    elif score >= 80:
        return "B"
    elif score >= 70:
        return "C"
    elif score >= 60:
        return "D"
    else:
        return "F"
```

```python
def analyze_grade(score):
    # 유효한 점수 범위인지 확인
    if score < 0 or score > 100:
        return "오류: 0~100 사이의 점수를 입력해주세요."
    
    # 등급 결정
    grade = grade_calculator(score)
    
    # 등급별 상세 피드백
    if grade == "A":
        if score >= 95:
            feedback = "탁월한 성과입니다! 최상위권 학생입니다."
        else:
            feedback = "아주 우수한 성적입니다. 조금만 더 노력하면 A+도 가능합니다."
    elif grade == "B":
        feedback = "우수한 성적입니다. 몇 가지 개념을 더 보완하면 A 등급도 가능합니다."
    elif grade == "C":
        feedback = "보통 수준의 성적입니다. 기본 개념을 다시 점검해보세요."
    elif grade == "D":
        feedback = "통과는 했지만 많은 개선이 필요합니다. 기초부터 다시 공부하세요."
    else: # F 등급
        if score >= 50:
            feedback = "아쉽게 과락입니다. 조금만 더 노력하면 통과할 수 있습니다."
        else:
            feedback = "기초 지식이 부족합니다. 선수 과목부터 다시 검토해보세요."
            
    # 결과 반환
    return f"점수: {score}, 등급: {grade}, 피드백: {feedback}"
```

```python
test_scores = [95, 85, 75, 65, 55, 45, 105, -5]

for score in test_scores:
    result = analyze_grade(score)
    print(result)
    print("-" * 50)
```

```python
def process_data(data):
    if data:
        if is_valid(data):
            if has_permission(data):
                # 실제 처리 코드
                return "처리 완료"
            else:
                return "권한 없음"
        else:
            return "유효하지 않은 데이터"
    else:
        return "데이터 없음"
```

```python
# 조기 반환 적용 (Early Return)
def process_data(data):
    if not data:
        return "데이터 없음"
        
    if not is_valid(data):
        return "유효하지 않은 데이터"
        
    if not has_permission(data):
        return "권한 없음"
        
    # 모든 조건을 통과한 경우에만 실행
    # 실제 처리 코드
    return "처리 완료"
```
