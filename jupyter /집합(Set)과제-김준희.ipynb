```python
# 사용자 관심사 데이터
raw_data = {
    "Alice": ["음악", "영화", "독서"],
    "Bob": ["스포츠", "여행", "음악"],
    "Charlie": ["프로그래밍", "게임", "영화"],
    "David": ["요리", "여행", "사진"],
    "Eve": ["프로그래밍", "독서", "음악"],
    "Frank": ["스포츠", "게임", "요리"],
    "Grace": ["영화", "여행", "독서"]
}

# 리스트를 집합으로 변환
users = {name: set(interests) for name, interests in raw_data.items()}
```

```python
target_user = "Alice"
target_interests = users[target_user]

common_interest_friends = []
no_common_interest_people = []

for name, interests in users.items():
    if name == target_user:
        continue
    
    # 1. 공통 관심사 확인 (교집합 연산)
    common = target_interests & interests
    if common:
        common_interest_friends.append((name, common))
    
    # 2. 공통 관심사 없음 확인 (서로소 연산)
    if target_interests.isdisjoint(interests):
        no_common_interest_people.append(name)

# --- 결과 출력 ---
print(f"[{target_user}]님을 위한 분석 리포트")
print("-" * 30)

print("1. 공통 관심사가 있는 추천 친구:")
for name, common in common_interest_friends:
    print(f"   - {name}: 공통 관심사 {common}")

print("\n2. 공통 관심사가 전혀 없는 사람:")
print(f"   - {', '.join(no_common_interest_people)}")
```
