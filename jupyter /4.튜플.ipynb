```python
# 튜플 메서드와 연산
numbers = (1, 2, 3, 2, 4, 2)

# count(): 특정 요소 개수 반환
print(numbers.count(2))

# index(): 특정 요소의 첫 번째 위치 반환
print(numbers.index(3)) 

# 튜플 연결
tuple1 = (1, 2, 3)
tuple2 = (4, 5, 6)
combined = tuple1 + tuple2 

# 튜플 반복
repeated = tuple1 * 3 

# 튜플 비교 (요소별 비교)
print((1, 2, 3) < (1, 3, 0)) 
```

```python
# 튜플 언패킹(Unpacking)
# 기본 언패킹
rgb = (255, 100, 50)
red, green, blue = rgb
print(f"Red: {red}, Green: {green}, Blue: {blue}")
```

```python
# 일부 요소만 언패킹
numbers = (1, 2, 3, 4, 5)
first, *middle, last = numbers
print(first)   
print(middle)  
print(last) 
```

```python
from collections import namedtuple

# 네임드 튜플 정의
Person = namedtuple('Person', ['name', 'age', 'city'])

# 생성 및 사용
person1 = Person('홍길동', 30, '서울')

# 인덱스로 접근
print(person1) 

# 필드명으로 접근 (가독성 높음)
print(person1.name) 
print(person1.age)  
print(person1.city) 

```
