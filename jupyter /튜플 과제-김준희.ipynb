0. 데이터 정의 및 초기화
분석할 데이터를 준비하고, 결과를 저장할 빈 바구니(딕셔너리)들을 만듭니다

```python
# 분석 대상 데이터
sales_data = [
    (2020, 1, "노트북", 1200, 100, "서울"),
    (2020, 1, "스마트폰", 800, 200, "부산"),
    (2020, 2, "노트북", 1200, 150, "서울"),
    (2020, 2, "스마트폰", 800, 250, "대구"),
    (2020, 3, "노트북", 1300, 120, "인천"),
    (2020, 3, "스마트폰", 850, 300, "서울"),
    (2020, 4, "노트북", 1300, 130, "부산"),
    (2020, 4, "스마트폰", 850, 350, "서울"),
    (2021, 1, "노트북", 1400, 110, "대구"),
    (2021, 1, "스마트폰", 900, 220, "서울"),
    (2021, 2, "노트북", 1400, 160, "인천"),
    (2021, 2, "스마트폰", 900, 270, "부산"),
    (2021, 3, "노트북", 1500, 130, "서울"),
    (2021, 3, "스마트폰", 950, 320, "대구"),
    (2021, 4, "노트북", 1500, 140, "부산"),
    (2021, 4, "스마트폰", 950, 370, "서울")
]

yearly_sales = {}      # 연도별 판매량
product_prices = {}    # 제품별 가격 (합계, 개수)
region_sales = {}      # 지역별 판매량
quarterly_revenue = {} # 분기별 매출 (가격 * 판매량)
```

```python
for row in sales_data:
    # 튜플 언패킹: 각 데이터를 의미 있는 변수명으로 추출
    year, quarter, product, price, quantity, region = row
    
    # 1. 연도별 판매량 합산
    yearly_sales[year] = yearly_sales.get(year, 0) + quantity
    
    # 2. 제품별 평균 가격을 위한 데이터 저장
    if product not in product_prices:
        product_prices[product] = [0, 0] # [총합계, 데이터개수]
    product_prices[product][0] += price
    product_prices[product][1] += 1
    
    # 3. 지역별 판매량 합산
    region_sales[region] = region_sales.get(region, 0) + quantity
    
    # 4. 분기별 매출 분석
    revenue = price * quantity
    quarterly_revenue[quarter] = quarterly_revenue.get(quarter, 0) + revenue
```

```python
for row in sales_data:
    # 튜플 언패킹: 각 데이터를 의미 있는 변수명으로 추출
    year, quarter, product, price, quantity, region = row
    
    # 1. 연도별 판매량 합산
    yearly_sales[year] = yearly_sales.get(year, 0) + quantity
    
    # 2. 제품별 평균 가격을 위한 데이터 저장 (리스트 활용)
    if product not in product_prices:
        product_prices[product] = [0, 0] # [총합계, 데이터개수]
    product_prices[product][0] += price
    product_prices[product][1] += 1
    
    # 3. 지역별 판매량 합산
    region_sales[region] = region_sales.get(region, 0) + quantity
    
    # 4. 분기별 매출 분석
    revenue = price * quantity
    quarterly_revenue[quarter] = quarterly_revenue.get(quarter, 0) + revenue
```

```python
print("-" * 15, "분석 결과 보고서", "-" * 15)

# 1. 연도별 판매량
print(f"1. 연도별 총 판매량: {yearly_sales}")

# 2. 제품별 평균 가격 계산 및 출력
print("\n2. 제품별 평균 가격:")
for prod, data in product_prices.items():
    avg = data[0] / data[1] # 평균 = 총합 / 개수
    print(f"   - {prod}: {avg:,.0f}원")

# 3. 최대 판매 지역 찾기
# max 함수의 key 인자를 활용해 가장 판매량이 높은 지역을 찾느나ㅣ딩.
max_region = max(region_sales, key=region_sales.get)
print(f"\n3. 최대 판매 지역: {max_region} ({region_sales[max_region]}개)")

# 4. 분기별 매출 합계 (분기 순서대로 정렬)
print("\n4. 분기별 매출 합계:")
for q in sorted(quarterly_revenue.keys()):
    print(f"   - {q}분기: {quarterly_revenue[q]:,.0f}원")

print("-" * 48)
```
