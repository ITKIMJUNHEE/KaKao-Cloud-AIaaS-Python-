```python
# 리스트 수정
fruits = ["사과", "바나나", "체리"]

# 요소 변경
fruits[1] = "블루베리"  # ['사과', '블루베리', '체리']

# 요소 추가
fruits.append("딸기")   # ['사과', '블루베리', '체리', '딸기']

# 특정 위치에 삽입
fruits.insert(1, "포도") # ['사과', '포도', '블루베리', '체리', '딸기']

# 리스트 확장
more_fruits = ["키위", "망고"]
fruits.extend(more_fruits) # ['사과', '포도', '블루베리', '체리', '딸기', '키위', '망고']

# 요소 제거 (첫 번째 일치하는 항목)
fruits.remove("체리")   # ['사과', '포도', '블루베리', '딸기', '키위', '망고']
```

```python
# 특정 위치 요소 제거 및 반환
removed = fruits.pop(2) # '블루베리' 제거 및 반환
print(removed)  # '블루베리'
print(fruits)   # ['사과', '포도', '딸기', '키위', '망고']

# 리스트 정렬
fruits.sort()   # ['망고', '딸기', '사과', '키위', '포도'] (알파벳 순)

# 리스트 역순으로 뒤집기
fruits.reverse() # ['포도', '키위', '사과', '딸기', '망고']
```

```python
# 리스트 길이
print(len(fruits))  # 5

# 요소 개수 세기
fruits.append("사과")
print(fruits.count("사과")) # 2 ('사과'가 2번 등장)

# 요소 위치 찾기
print(fruits.index("키위")) # 3 ('키위'의 위치)

# 리스트 내용 비우기
fruits.clear()  # []
```

```python
# enumerate 객체 변환 예시
letters = ['a', 'b', 'c']
enum_obj = enumerate(letters)
print(list(enum_obj)) # [(0, `a`), (1, `b`), (2, `c`)]
```

```python
# enumerate 함수 활용하기
# 리스트 요소와 인덱스를 함께 순회할 때 유용
fruits = ["사과", "바나나", "체리", "딸기"]
for index, fruit in enumerate(fruits):
    print(f"{index}번: {fruit}") # '0번: 사과', '1번: 바나나' 등 출력

# 시작 인덱스 지정
for index, fruit in enumerate(fruits, start=1):
    print(f"{index}번: {fruit}") # '1번: 사과', '2번: 바나나' 등 출력
```

```python
# 리스트의 특정 값 위치 찾기
colors = ["빨강", "파랑", "초록", "파랑", "노랑"]
blue_indices = [i for i, color in enumerate(colors) if color == "파랑"]
print(f"파랑의 위치: {blue_indices}") # [1, 3]

# enumerate를 사용하지 않으면?
fruits = ["사과", "바나나", "체리"]
for i in range(len(fruits)):
    print(f"{i}번: {fruits[i]}")
```

```python
tasks = [] # 할 일을 담을 빈 리스트

# 1. 할 일 추가 함수
def add_task(task_name):
    tasks.append(task_name)

# 2. 할 일 완료(삭제) 함수
def complete_task(index):
    if 0 <= index < len(tasks):
        removed = tasks.pop(index)
        print(f"['{removed}'] 항목을 완료했습니다!")

# 3. 할 일 목록 조회
def view_tasks():
    if tasks:
        print("\n===== 할 일 목록 =====")
        for i, task in enumerate(tasks):
            print(f"{i+1}. {task}")
        print("======================")
    else:
        print("\n할 일이 없습니다.")

# --- 프로그램 실행 예시 ---
add_task("파이썬 공부하기")
add_task("장보기")
add_task("운동하기")

view_tasks() # 첫 번째 목록 확인

complete_task(1) # 인덱스 1번("장보기") 완료 처리

view_tasks() # 바뀐 목록 확인
```
