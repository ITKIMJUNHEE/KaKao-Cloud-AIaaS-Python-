```python
# 리스트 수정
fruits = ["사과", "바나나", "체리"]

# 요소 변경
fruits[1] = "블루베리"  # ['사과', '블루베리', '체리']

# 요소 추가
fruits.append("딸기")   # ['사과', '블루베리', '체리', '딸기']

# 특정 위치에 삽입
fruits.insert(1, "포도") # ['사과', '포도', '블루베리', '체리', '딸기']

# 리스트 확장
more_fruits = ["키위", "망고"]
fruits.extend(more_fruits) # ['사과', '포도', '블루베리', '체리', '딸기', '키위', '망고']

# 요소 제거 (첫 번째 일치하는 항목)
fruits.remove("체리")   # ['사과', '포도', '블루베리', '딸기', '키위', '망고']
```

```python
# 특정 위치 요소 제거 및 반환
removed = fruits.pop(2) # '블루베리' 제거 및 반환
print(removed)  # '블루베리'
print(fruits)   # ['사과', '포도', '딸기', '키위', '망고']

# 리스트 정렬
fruits.sort()   # ['망고', '딸기', '사과', '키위', '포도'] (알파벳 순)

# 리스트 역순으로 뒤집기
fruits.reverse() # ['포도', '키위', '사과', '딸기', '망고']
```

```python
# 리스트 길이
print(len(fruits))  # 5

# 요소 개수 세기
fruits.append("사과")
print(fruits.count("사과")) # 2 ('사과'가 2번 등장)

# 요소 위치 찾기
print(fruits.index("키위")) # 3 ('키위'의 위치)

# 리스트 내용 비우기
fruits.clear()  # []
```

```python
# enumerate 함수 활용하기
# 리스트 요소와 인덱스를 함께 순회할 때 유용
fruits = ["사과", "바나나", "체리", "딸기"]
for index, fruit in enumerate(fruits):
    print(f"{index}번: {fruit}") # '0번: 사과', '1번: 바나나' 등 출력

# 시작 인덱스 지정
for index, fruit in enumerate(fruits, start=1):
    print(f"{index}번: {fruit}") # '1번: 사과', '2번: 바나나' 등 출력
```

```python
# 할 일 목록을 저장할 글로벌 리스트
tasks = []
```

```python
# 1. 할 일 추가 함수
def add_task(task_name):
    tasks.append(task_name)
    print(f"새로운 할 일 '{task_name}'이(가) 추가되었습니다.")
```

```python
# 2. 할 일 목록 조회 함수
def view_tasks():
    if tasks:
        print("\n===== 할 일 목록 =====")
        for i, task in enumerate(tasks):
            # 사용자 편의를 위해 1번부터 번호를 매겨 출력합니다.
            print(f"{i+1}. {task}")
        print("======================")
    else:
        print("\n할 일이 없습니다.")
```

```python
# 3. 할 일 완료(삭제) 함수
def complete_task(index):
    # 인덱스 범위가 유효한지 확인합니다.
    if 0 <= index < len(tasks):
        # 0부터 시작하는 리스트 인덱스에 맞춰 요소를 제거합니다.
        completed = tasks.pop(index)
        print(f"✅ '{completed}' 할 일을 완료했습니다!")
    else:
        print("잘못된 번호입니다.")
```

```python
# 할 일 목록 조회
def view_tasks():
    if tasks:
        print("\n===== 할 일 목록 =====")
        for i, task in enumerate(tasks):
            print(f"{i+1}. {task}")
        print("======================")
    else:
        print("할 일이 없습니다.")
```

```python
# --- 프로그램 실행 예시 ---
add_task("파이썬 공부하기")  # index 0
add_task("장보기")            # index 1
add_task("운동하기")          # index 2
```

```python
view_tasks()  # 전체 목록 출력

complete_task(1) # "장보기" 완료 및 삭제

view_tasks()  # 갱신된 목록 출력
```

```python
# 얕은 복사의 문제점 예시
original = [1, 2, [3, 4]]
shallow = original.copy()

# 내부 리스트 요소 변경
shallow[2][0] = 30

print(original) # [1, 2, [30, 4]] - 원본도 변경됨!
print(shallow) # [1, 2, [30, 4]]
```

```python
import copy

original = [1, 2, [3, 5]]
deep = copy.deepcopy(original)

# 내부 리스트 요소 변경
deep[2] = 30

print(original) # [1, 2, [3, 5]] - 원본은 변경되지 않음
print(deep)     # [1, 2, [5, 12]]
```
