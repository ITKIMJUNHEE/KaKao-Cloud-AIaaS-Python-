실습 1_기본

```python
def greet(name, time_of_day="아침"):
    greetings = {
        "아침": "좋은 아침입니다",
        "점심": "맛있는 점심 드세요",
        "저녁": "편안한 저녁 되세요"
    }
    return f"{greetings[time_of_day]}, {name}님!"

print(greet("dahan"))

```

변스 스코프 (스코프 규칙)

```python
x = "전역"

def  outer ():
    x = "외부"
    def  inner():
        x = "내부"
        print(x)  # 내부 출력
    inner()
    print(x)      # 외부 출력
outer()
print(x)          # 전역 출력   
```

변수 스코프 ( global 키워드)

```python
count = 0   
def increment():
    global count  
    count += 1

increment()
print(count)  # 1
```

변수 사용 연습

```python
student_name = "김철수"
student_age = 20
student_height = 175.5
is_enrolled = True

# 정보 출력
print("=== 학생 정보 ===")
print(f"이름: {student_name}")
print(f"나이: {student_age}세")
print(f"키: {student_height}cm")
print(f"등록 여부: {'등록' if is_enrolled else '미등록'}")
```

변수 사용 연습 실습 2: 수학 연산

```python
x = 10
y = 3

# 기본 연산
addition = x + y
subtraction = x - y
multiplication = x * y
division = x / y
remainder = x % y

print("\n=== 수학 연산 ===")
print(f"덧셈: {x} + {y} = {addition}")
print(f"뺄셈: {x} - {y} = {subtraction}")
print(f"곱셈: {x} * {y} = {multiplication}")
print(f"나눗셈: {x} / {y} = {division}")
print(f"나머지: {x} % {y} = {remainder}")
```

변수 사용 연습 실습 3: 문자열연산

```python
# 실습 3: 문자열 연산
first_name = "홍"
last_name = "길동"
full_name = last_name + first_name
name_length = len(full_name)

print("\n=== 문자열 연산 ===")
print(f"이름: {full_name}")
print(f"이름 길이: {name_length}")
```
